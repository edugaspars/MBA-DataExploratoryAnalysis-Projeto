- Distribui√ß√µes Individuais:

Distribui√ß√µes Assim√©tricas: A maioria das vari√°veis apresenta distribui√ß√µes assim√©tricas, com algumas vari√°veis tendo uma cauda mais longa √† direita, como <Strong>proline e total_phenols</Strong>. Essa informa√ß√£o pode ser √∫til para a sele√ß√£o de modelos de regress√£o ou classifica√ß√£o.

```python
def empirical_rule(data):
    mean = np.mean(data)
    std = np.std(data)
    within_1_std = ((data > (mean - std)) & (data < (mean + std))).mean() * 100
    within_2_std = ((data > (mean - 2 * std)) & (data < (mean + 2 * std))).mean() * 100
    within_3_std = ((data > (mean - 3 * std)) & (data < (mean + 3 * std))).mean() * 100
    return within_1_std, within_2_std, within_3_std

def chebyshev_rule(data, k):
    mean = np.mean(data)
    std = np.std(data)
    lower_bound = mean - k * std
    upper_bound = mean + k * std
    within_k_std = ((data > lower_bound) & (data < upper_bound)).mean() * 100
    return within_k_std
```

1. Fun√ß√£o empirical_rule:

Essa fun√ß√£o implementa a regra emp√≠rica, que afirma que para distribui√ß√µes normais, aproximadamente 68% dos dados ficam dentro de 1 desvio padr√£o da m√©dia, 95% dentro de 2 desvios padr√£o e 99,7% dentro de 3 desvios padr√£o.
A fun√ß√£o calcula essas porcentagens para o conjunto de dados fornecido.
Ela calcula a m√©dia e o desvio padr√£o dos dados usando np.mean e np.std.
Em seguida, usa opera√ß√µes booleanas para encontrar os dados que est√£o dentro de cada intervalo de desvio padr√£o.
O resultado √© retornado como uma tupla contendo as porcentagens dentro de 1, 2 e 3 desvios padr√£o.


2. Fun√ß√£o chebyshev_rule:

Essa fun√ß√£o implementa a regra de Chebyshev, que fornece um limite inferior para a porcentagem de dados que ficam dentro de um determinado n√∫mero de desvios padr√£o da m√©dia.
Ela aceita um conjunto de dados e um valor k que representa o n√∫mero de desvios padr√£o.
Ela calcula o limite inferior e superior para o intervalo k de desvios padr√£o da m√©dia.
Ent√£o, usa opera√ß√µes booleanas para encontrar os dados dentro desse intervalo e calcula a porcentagem desses dados.

```python
# Dividindo os dados para regress√£o
from sklearn.metrics import confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(df_wine.drop('TARGET', axis=1), df_wine['TARGET'], test_size=0.3, random_state=42)

# Modelo de Regress√£o Log√≠stica com varia√ß√µes param√©tricas
parameters = [
    {'penalty': 'l2', 'C': 1.0, 'solver': 'lbfgs', 'max_iter': 100},  # Default parameters
    {'penalty': 'l2', 'C': 0.1, 'solver': 'lbfgs', 'max_iter': 100},  # Lower regularization
    {'penalty': 'l2', 'C': 10.0, 'solver': 'lbfgs', 'max_iter': 100}  # Higher regularization
]
results = []

# Avaliando os modelos com diferentes par√¢metros
for param in parameters:
    model = LogisticRegression(**param)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)[:, 1]
    cm = confusion_matrix(y_test, predictions)
    score = model.score(X_test, y_test)
    print("Coef (A):", model.coef_)
    print("Intercept (B):", model.intercept_)

    plt.figure(figsize=(9,9))
    sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
    plt.ylabel('Actual label');
    plt.xlabel('Predicted label');
    all_sample_title = 'Accuracy Score: {0}'.format(score)
    plt.title(all_sample_title, size = 15)
```

```python
# Configura√ß√µes de C para a regress√£o log√≠stica com valida√ß√£o cruzada
C_values = [0.1, 1.0, 10.0]

# Criando o modelo com valida√ß√£o cruzada
model_cv = LogisticRegressionCV(Cs=C_values, cv=5, random_state=42, max_iter=10000)
model_cv.fit(X_train, y_train)

# Coletando m√©dias dos scores da valida√ß√£o cruzada para cada C
mean_accuracy = np.mean(model_cv.scores_[1], axis=0)

# Plotando Acur√°cia por configura√ß√£o de C
plt.figure(figsize=(10, 5))
plt.plot(C_values, mean_accuracy, marker='o', linestyle='-', color='blue')
plt.title('Acur√°cia M√©dia por Valor de C')
plt.xlabel('Valor de C')
plt.ylabel('Acur√°cia M√©dia')

plt.tight_layout()
plt.show()
```

Isso ajuda a encontrar o valor de C que oferece o melhor equil√≠brio entre complexidade do modelo e desempenho

O valor de C = 1.0 parece ser o ideal, pois maximiza a acur√°cia m√©dia sem diminuir o desempenho com regulariza√ß√£o excessiva ou flexibilidade excessiva.
